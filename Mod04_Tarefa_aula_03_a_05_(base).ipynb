## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui
#usando a media de acerto de cada materia e calculando o percentual
ac_mat = (df_mat['Qt_acertos'].mean()/24)*100
print(f'A media de acertos de matematica {round(ac_mat,2)}%.')
ac_por = (df_por['Qt_acertos'].mean()/18)*100
print(f'A media de acertos de portugues {round(ac_por,2)}%.')
ac_geo = (df_geo['Qt_acertos'].mean()/8)*100
print(f'A media de acertos de geografia {round(ac_geo,2)}%.')
ac_ing = (df_ing['Qt_acertos'].mean()/8)*100
print(f'A media de acertos de ingles {round(ac_ing,2)}%.')
ac_his = (df_his['Qt_acertos'].mean()/8)*100
print(f'A media de acertos de historia {round(ac_his,2)}%.')
ac_fis = (df_fis['Qt_acertos'].mean()/12)*100
print(f'A media de acertos de fisica {round(ac_fis,2)}%.')
ac_qui = (df_qui['Qt_acertos'].mean()/12)*100
print(f'A media de acertos de quimica {round(ac_qui,2)}%.')
````

````python
# 2) Seu código aqui
#somando todos df para termos o total de acerto de todos alunos, mudei o nome das colunas em cada meteria.
np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_mat'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_fis'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_qui'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_por'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_geo'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_ing'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_hist'])

df = pd.concat([df_mat, df_fis, df_qui, df_por, df_geo, df_ing, df_his], axis=1)

df ['total'] = df['Qt_mat'] + df['Qt_fis'] + df['Qt_qui'] + df['Qt_por'] + df['Qt_geo'] + df['Qt_ing'] + df['Qt_hist']

df
````

````python
# 3) Seu código aqui
# percentual geral de cada aluno com o total de 90 questoes
df ['media'] = (df['total']/90)*100
print(df)
````

````python
# 4) Seu código aqui

#Alunos que passaram
df_passaram = df.loc[df['total'] >= 45]
#quantidade de alunos que passaram
len(df_passaram)
````
